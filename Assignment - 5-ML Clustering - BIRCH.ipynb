```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import Birch # Birch Clustering
# Load the dataset
dataset = pd.read_csv('Mall_Customers.csv') 
# Select the features for clustering. 
X = dataset.iloc[:, 3:5].values
birch_model = Birch(n_clusters=5)  # Specify the number of clusters
y_pred = birch_model.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
for cluster_id in np.unique(y_pred):
	plt.scatter(X[y_pred == cluster_id, 0], X[y_pred == cluster_id, 1], label=f'Cluster {cluster_id + 1}')
plt.title('Clusters using Birch Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
